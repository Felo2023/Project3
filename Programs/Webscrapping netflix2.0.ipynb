# Netflix worldwide top 10 scrap

```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
import pandas as pd
import time
```

```python
# Set up Splinter
browser = Browser('chrome')
```

```python
# Visit Netflix top 10 global site
url = 'https://www.netflix.com/tudum/top10?utm_source=google&utm_medium=cpc&utm_campaign=september&utm_id=09&gclid=Cj0KCQjw06-oBhC6ARIsAGuzdw3Kzz8edhpv0YSQiLhIxxZ3BIhZlFJC2qPqaclHvQAPhB-ejtGVHboaAkLxEALw_wcB'
browser.visit(url)
```

```python
df_master=pd.DataFrame
df=pd.DataFrame
for i in range(115):
    current_url=browser.url
    print(current_url)
    # Find the table
    html = browser.html
    html_soup = soup(html, 'html.parser')
    table = html_soup.find('table', class_='table-fixed')

    # Store the table
    data=table.find('tbody')
    objects = data.find_all('tr')
    table1=[]

    # Loop through the rows
    for object_ in objects:
        row=[]
        #row_heading = row.find('th').text
        object_data = object_.find_all('td')
        for cell in object_data:
            data1=cell.text
            row.append(data1)
        table1.append(row)

    if len(row)<5:
        headers=['No','movie','week_at_Top10', 'hours_seen']
    else:
        headers=['No','movie','week_at_Top10', 'hours_seen', 'duration','views']

    if i==0:    
        df_master=pd.DataFrame(table1,columns=headers)
    else:
        df=pd.DataFrame(table1,columns=headers)
        df_master=pd.concat([df_master,df],sort=False)
    browser.find_by_xpath('/html/body/div[1]/div/main/section[2]/div/div[3]/div/div/button[1]/img').click()
    time.sleep(1)

```

```python
df_master['hours_seen']=df_master['hours_seen'].str.replace(r'(\,)', r'',regex=True)
df_master
```

```python
df_master['views']=df_master['views'].str.replace(r'(\,)', r'',regex=True)
df_master.head(20)
```
