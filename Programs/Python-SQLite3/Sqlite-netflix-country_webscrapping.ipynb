```python
# Import the dependencies.
from pathlib import Path
from sqlalchemy import create_engine, text
import sqlite3
import csv
```

```python
    # Import webscrapping csv and extract data
    country='united-kingdom'
    with open(f'../../webscrapping/Netflix_all_top10_{country}.csv', 'r') as fin:
        dr = csv.DictReader(fin)
        country_top10 = [(i['No'],i['movie'],i['week_at_Top10']) for i in dr]
        
```

```python
    # Connect to SQLite
    sqliteConnection = sqlite3.connect('Netflix_DB.sqlite')
    cursor = sqliteConnection.cursor()
```

```python
# Create countries table
cursor.execute(f'''CREATE TABLE united_kingdom (
    ID INTEGER PRIMARY KEY AUTOINCREMENT,
    No int,
    movie varchar (150),
	week_at_Top10 int
);''')
```

```python
    # Insert data into table
    cursor.executemany(
        f"insert into united_kingdom (No, movie, week_at_Top10) VALUES (?, ?, ?);", country_top10)
```

```python
    # Show student table
    cursor.execute(f'select * from united_kingdom ;')
```

```python
    # View result
    result = cursor.fetchall()
    print(result)
```

```python
    # Commit work and close connection
    sqliteConnection.commit()
    
```
