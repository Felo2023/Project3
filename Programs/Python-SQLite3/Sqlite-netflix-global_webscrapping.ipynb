```python
# Import the dependencies.
from pathlib import Path
from sqlalchemy import create_engine, text
import sqlite3
import csv
```

```python
    # Import webscrapping csv and extract data
    with open('../../webscrapping/Netflix_all_top10_worldwide.csv', 'r') as fin:
        dr = csv.DictReader(fin)
        global_top10 = [(i['No'],i['movie'],i['week_at_Top10'],i['hours_seen'],i['duration'],i['views']) for i in dr]
        
```

```python
    # Connect to SQLite
    sqliteConnection = sqlite3.connect('Netflix_DB.sqlite')
    cursor = sqliteConnection.cursor()
```

```python
cursor.execute('DROP TABLE global1')
```

```python
# Create countries table
cursor.execute('''CREATE TABLE global1(
	ID INTEGER PRIMARY KEY AUTOINCREMENT,
	No int,
    movie varchar (150),
	week_at_Top10 int,
	hours_seen decimal,
	duration time,
	views decimal
);''')
```

```python
    # Insert data into table
    cursor.executemany(
        "insert into global1 (No, movie, week_at_Top10, hours_seen, duration, views) VALUES (?, ?, ?, ?, ?, ?);", global_top10)
```

```python
    # Show student table
    cursor.execute('select * from global1;')
```

```python
    # View result
    result = cursor.fetchall()
    print(result)
```

```python
    # Commit work and close connection
    sqliteConnection.commit()
    cursor.close()
```
